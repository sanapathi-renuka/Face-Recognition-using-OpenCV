```python
import cv2
from matplotlib import pyplot as  plt
```

```python
img=cv2.imread('face3.jpg')
```

```python
img_gray2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img_rgb2=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
haarcascade_data=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
found=haarcascade_data.detectMultiScale(img_gray2,minSize=(20,20))
```

```python
#if no sign dont do anything
amount_found=len(found)
if amount_found!=0:
    #for multiple signs
    for(x,y,width,height) in found:
        cv2.rectangle(img_rgb2,(x,y),(x+height,y+width),(0,255,0),5)
```

```python
plt.subplot(1,1,1)
plt.imshow(img_rgb2)
plt.show()
```
